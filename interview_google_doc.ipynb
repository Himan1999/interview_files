```python
from bs4 import BeautifulSoup, Tag
import requests as req


Webpage = open("doc_google/PyTorch Consulting/PyTorchConsulting.html","r")
webpagehtml = Webpage.read()
lookupwebpage = BeautifulSoup(webpagehtml,'html.parser')
print(lookupwebpage.prettify())
```

```python

new_tag = lookupwebpage.new_tag("h3")
for i in lookupwebpage.find_all("span", attrs= {'class':"c9"}):
    new_tag.string = i.get_text()
    i.replace_with(new_tag)
```

```python

new_tag = lookupwebpage.new_tag("h3")
for i in lookupwebpage.find_all("span", attrs= {'style':"""font-size:16.5pt;mso-fareast-font-family:"Times New Roman";mso-bidi-font-family:
Calibri;mso-bidi-theme-font:minor-latin;color:#0070C0;mso-fareast-language:
EN-GB"""}):
    new_tag.string = i.get_text()
    i.replace_with(new_tag)
```
